```python
train_data = pd.read_csv("data/train.csv")
test_data = pd.read_csv("data/test.csv")

train_data['Age'].fillna(train_data['Age'].median(), inplace=True)
test_data['Age'].fillna(test_data['Age'].median(), inplace=True)
```

```python
from sklearn.model_selection import train_test_split

y = train_data["Survived"]
features = ["Pclass", "Sex", "SibSp", "Parch", "Age", "Fare"]
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

# Dividir train_data em treino e validação (80/20)
train_X, val_X, train_y, val_y = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_absolute_error, accuracy_score

# 1. Treinar com dados de treino
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(train_X, train_y)

# 2. Validar com dados de validação
val_predictions = model.predict(val_X)
val_mae = mean_absolute_error(val_y, val_predictions)
val_accuracy = accuracy_score(val_y, val_predictions)

print(f"Validation MAE: {val_mae:.4f}")
print(f"Validation Accuracy: {val_accuracy:.4f}")

# 3. Retreinar com TODOS os dados para submissão final
model.fit(X, y)
predictions = model.predict(X_test)

# 4. Criar arquivo de submissão
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
